###A segmentation by Success_of_the_session column

```python
query32= """
WITH Success_table AS (SELECT u.user_id, COUNT(s.session_id) AS num_sessions,
CASE WHEN s.cancellation='False' AND s.flight_booked='False' AND s.hotel_booked='False' THEN 'no_booked'
WHEN s.cancellation='True' AND s.flight_booked='True' AND s.hotel_booked='True'  THEN 'booked_and_cancelled'
ELSE 'Yes'
END AS Success_of_the_session
FROM users u
JOIN sessions s
USING (user_id)
WHERE s.session_start>'2023-01-04'
GROUP BY u.user_id, Success_of_the_session

ORDER BY Success_of_the_session)

SELECT Success_of_the_session,COUNT(user_id) AS num_users
FROM Success_table
GROUP BY Success_of_the_session

"""
display(pd.read_sql(sa.text(query32),connection))
```

###Let us look at what happens to users with less than 3 sessions.

```python
query33= """
WITH Success_table AS (SELECT u.user_id, COUNT(s.session_id) AS num_sessions,
CASE WHEN s.cancellation='False' AND s.flight_booked='False' AND s.hotel_booked='False' THEN 'no_booked'
WHEN s.cancellation='True' AND s.flight_booked='True' AND s.hotel_booked='True'  THEN 'booked_and_cancelled'
ELSE 'Yes'
END AS Success_of_the_session
FROM users u
JOIN sessions s
USING (user_id)
WHERE s.session_start>'2023-01-04'
GROUP BY u.user_id, Success_of_the_session
HAVING COUNT(s.session_id)<3
ORDER BY Success_of_the_session)

SELECT Success_of_the_session,COUNT(user_id) AS num_users
FROM Success_table
GROUP BY Success_of_the_session

"""
display(pd.read_sql(sa.text(query33),connection))
```

###Let us look at what happens to users with more than or equal to 3 sessions.

```python
query34= """
WITH Success_table AS (SELECT u.user_id, COUNT(s.session_id) AS num_sessions,
CASE WHEN s.cancellation='False' AND s.flight_booked='False' AND s.hotel_booked='False' THEN 'no_booked'
WHEN s.cancellation='True' AND s.flight_booked='True' AND s.hotel_booked='True'  THEN 'booked_and_cancelled'
ELSE 'Yes'
END AS Success_of_the_session
FROM users u
JOIN sessions s
USING (user_id)
WHERE s.session_start>'2023-01-04'
GROUP BY u.user_id, Success_of_the_session
HAVING COUNT(s.session_id)>=3
ORDER BY Success_of_the_session)

SELECT Success_of_the_session,COUNT(user_id) AS num_users
FROM Success_table
GROUP BY Success_of_the_session

"""

display(pd.read_sql(sa.text(query34),connection))
```

###Conclusion: Users with more than 3 sessions do not cancel or do not book.
